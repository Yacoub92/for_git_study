```python
import pandas as pd
import seaborn as sns
import numpy as np

# Read the CSV file
df = pd.read_csv('train.csv')

# Identify the target variable
target_variable = 'SalePrice'

# Display the distribution of the target variable using seaborn's histogram plot
sns.histplot(df[target_variable], kde=True)

# Calculate kurtosis and skewness of the target variable
print(f"Kurtosis: {df[target_variable].kurtosis()}")
print(f"Skewness: {df[target_variable].skew()}")

# Apply logarithmic transformation to the target variable
log_target_variable = np.log1p(df[target_variable])

# Display the distribution of the logarithmically transformed target variable using seaborn's histogram plot
sns.histplot(log_target_variable, kde=True)

# Calculate kurtosis and skewness of the logarithmically transformed target variable
print(f"Kurtosis (after logarithmic transformation): {log_target_variable.kurtosis()}")
print(f"Skewness (after logarithmic transformation): {log_target_variable.skew()}")
```

```python
import pandas as pd
import seaborn as sns



# Read the CSV file

df = pd.read_csv(r'C:\Users\LENOVO THINKPAD W550\Desktop\general_practice\train.csv')

# Identify the target variable
target_variable = 'SalePrice'

# Create a correlation matrix
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr_matrix)

# Select 10 features highly correlated with the target variable
corr_with_target = corr_matrix[target_variable]
high_corr_features = corr_with_target[abs(corr_with_target) > 0.5].sort_values(ascending=False).index[:10]

# Create a heatmap of the correlation matrix for the selected 10 features
sns.heatmap(df[high_corr_features].corr())



# Find 3 combinations of highly correlated features
corr_matrix_10 = df[high_corr_features].corr()
high_corr_combinations = [(i, j) for i in corr_matrix_10.columns for j in corr_matrix_10.columns if i != j and corr_matrix_10.loc[i,j] > 0.7]
print(high_corr_combinations)
```
